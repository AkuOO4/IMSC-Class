```r
install.packages("igraph")
```

```r
library(igraph)

#making adjecency list
adj_list<-list(
  c(2,4,4),
  c(1,2),
  c(1,3,5),
  c(3,5,1),
  c(4,1)
)
#making a graph from the adj_list
g<-graph_from_adj_list(adj_list)
plot(g)
```

```r
#adding edges
g<-add_edges(g,c(1,5,3,1))
plot(g)
```

```r
#doing dfs on g Starting from 2
dfs(g,2)
```
